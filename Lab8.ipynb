## DHCP Discover

```python
mac = get_if_hwaddr(conf.iface)

dhcp_discover = (
    Ether(dst="ff:ff:ff:ff:ff:ff") /
    IP(src="0.0.0.0", dst="255.255.255.255") /
    UDP(sport=68, dport=67) /
    BOOTP(chaddr=mac) / 
    DHCP(options=[("message-type", "discover"), "end"])
    )

sendp(dhcp_discover, iface=conf.iface)
```
