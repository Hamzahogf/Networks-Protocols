```python
from scapy.all import *
```

## DNS Packet

```python
packet = IP(dst="8.8.8.8")/UDP(sport=RandShort(), dport=53)/DNS(rd=1, qd=DNSQR(qname="www.google.com", qtype = "A")) # default qtype in DNSQR is A

response = sr1(packet, verbose=0)
response.show()
```
